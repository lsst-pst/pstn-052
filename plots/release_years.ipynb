```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
import rubin_sim.maf.metricBundles as metricBundles
import rubin_sim.maf.slicers as slicers
import rubin_sim.maf.db as db
import rubin_sim.maf.utils as utils
import rubin_sim.maf.metrics as metrics
import sqlite3
from rubin_sim.maf.metrics.snNSNMetric import SNNSNMetric
import healpy as hp
import glob
from pltrollspot import spot_inspect
import pandas as pd
```

```python
dbfiles = glob.glob('baseline*10yrs.db')
dbfiles.extend(glob.glob('rolling*10yrs.db'))
dbfiles.sort()
outDir = 'yearly_release'
```

```python
sqls = ['night < %i and note not like "DD%%"' % 365, 'night<%i and note not like "DD%%"' % (365.25*3), 'night < %i and note not like "DD%%"' % (365.25*7)]

titles=['Y1', 'Y3', 'Y7']
mins = [0, 25,50 ]
maxes = [140, 475, 800]
for sql, colormin, colormax, title in zip(sqls,mins,maxes,titles):
    for filename in dbfiles:
        opsimdb = db.OpsimDatabase(filename)
        runname = filename.replace('_10yrs.db', '')

        resultsDb = db.ResultsDb(outDir=outDir)
        bundleList = []
        metric = metrics.CountMetric(col='night', units='N obs', metricName='Count')
        slicer = slicers.HealpixSlicer()

        plotDict = {'colorMin':colormin, 'colorMax': colormax, 'title': runname+', %s' % title}
        
        bundleList.append(metricBundles.MetricBundle(metric, slicer,sql,runName=runname, plotDict=plotDict))


        bundleDict = metricBundles.makeBundlesDictFromList(bundleList)
        bgroup = metricBundles.MetricBundleGroup(bundleDict, opsimdb, outDir=outDir, resultsDb=resultsDb)
        bgroup.runAll()
        bgroup.plotAll(closefigs=False)
```
